```python
from pathlib import Path
import csv

 #set the file path

count = 0
total = 0
average = 0
minimum = 0
maximum = 0
delta_min = 0
delta_max = 0
delta = 0
previous_row_value = 0
Best_month = ''
Worst_month = ''
output = 0

csvpath = Path('C:/Users/Mirek/Python excercises/USYD-VIRT-FIN-PT-03-2022-U-LOL-main/USYD-VIRT-FIN-PT-03-2022-U-LOL-main/02-Python/Homework/PyBank/Resources/budget_data.csv')

    #open the input path as file object, contextual syntax:

with open(csvpath, 'r') as csvfile:
    print(type(csvfile))
       
    csvreader = csv.reader(csvfile, delimiter=',')
   
    print(type(csvreader))
    
    header = next(csvreader)
    print(f'{header} <-----HEADER')
    
  
    for row in csvreader: 
        print(row) 
        
        if count != 0:
            delta = int(row[1]) - previous_row_value
            
        if delta > delta_max:
            delta_max = delta
            Best_month = row[0]
        
        if delta < delta_min:
            delta_min = delta
            Worst_month = row[0] 
            
        total += int(row[1])
        count += 1
        
        if minimum == 0:
            minimum = int(row[1])
        elif int(row[1]) < minimum:
            minimum = int(row[1])
        elif int(row[1]) > maximum:
            maximum = int(row[1])
        
        previous_row_value = int(row[1])

    # Calculate the average

average = round(total / count, 2)


# Print out the summary statistics

with open(output, 'w') as csvfile:
    print(type(csvfile))
       
    csvwriter = csv.reader(csvfile, delimiter=',')

print("---------Summary Statistics----------")
print(f"Number of Total Months: {count}")
print(f"Total Profits/Losses: ${total}")
print(f"Daily Average: {average}")
print(f"Worst Loss month: ${minimum}")
print(f"Best Gain month: ${maximum}")
print(f"Greatest loss: {Worst_month} ${delta_min}")
print(f"Greated gain: {Best_month} ${delta_max}")
```
